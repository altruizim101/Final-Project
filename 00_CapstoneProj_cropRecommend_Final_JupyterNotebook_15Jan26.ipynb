```python
# -- Capstone Project, 15 January 2026 -- 
# -- Jupyter Notebook, Python Code: implememntation of Crop Recommnedation system using Machine Learning.

# -- Start of Python Code --
# 
# import required libraries, models
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.datasets import fetch_openml

from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score, cross_val_predict
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.svm import SVC, SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingRegressor

from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.inspection import permutation_importance

# make directory "plots" for saving results (plots and graphs).
import os
os.makedirs("plots", exist_ok=True)

# Ignore any warnings from code execution (will not impact results)
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["figure.figsize"] = (8, 6)

# -- Load data: data was obtained/ downloaded from Kaggle platform --
# 
data_df = set_ratings = pd.read_csv("/Users/m.jalloh/Downloads/00_FinalApp_CropRecommend/00_FinalApp_CropRecommend.csv")

# -- Identify "features" and "target" in dataset: For the crop recommendation, the target is the "label" column.
# 
X_features = data_df.drop(["label"], axis = 1)
y_target = data_df["label"]

# -- Understand data: 
# display data with Freatures (columns) and Samples (rows); how many Features? How many samples? ...
# what are the Features?, Is dataset "balanced" or not?, Is data with missing values? ...
# 
# -- Note: This data is a balanced dataset. Also with no missing values.
# 
data_df.info()
data_df.head()
print(data_df.shape)
print(data_df.columns)
# samples of a Feature
data_df["label"].sample(5)

# Differnt crop types: total of 22 different crops observed in target "label" column.
unique_count = data_df["label"].nunique()

# -- Dataset split: prepare Train & Test datasets
# splitting data into train and test sets is essential for building ML models. The "train" set is used to teach the model about 
# patterns and relationships between input features and target class. The "test" set evaluates how well the model performs on new,
# unseen data.
# 
X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size = 0.3, random_state = 42)

# --- Data Visualization --
# This is an essential step before ML modeling because it helps in understanding the structure & quality of the data. 
# Visualizing features in the data allows us to identify patterns & relationships between variables that may influence model performance. 
# It also helps detect missing values, outliers, and data imbalance that could negatively affect modeling process.
# This step aids in reducing errors, improving model design, and increasing confidence that the ML model is built on well-understood 
# data.

data_df_num = data_df.select_dtypes(include = "number")

corr_matrix = data_df_num.corr()
plt.figure(figsize = (8, 6))
sns.heatmap(corr_matrix, annot = True, linewidth = 0.8)
plt.title("Correlation Heatmap of Features for Crop Recommendation")
plt.show()

# -- Visualization with Pairplot --
# PairPlot: a quick, comprehensive view of relationships between the Features in dataset.
# creates scatterplots for every pair of Features, revealing correlations and potential patterns that 
# might influence predictions.
# 
pairgrid = sns.pairplot(data_df_num)
pairgrid.fig.suptitle("Pairplot of Features for Crop Recommendation")
pairgrid.fig.savefig("CapstoneProject_Pairplot.pdf", bbox_inches = "tight")
plt.show()

# -- Visualization: Correlation & Covariance matrix Heatmaps --
# Covariance is a measure of how two features in the data vary together.
# Correlation helps identify which Feature in the data is strongly correlated to the Target.
# 
ax = sns.heatmap(corr_matrix, annot = True, linewidth = 0.8)
ax.figure.savefig('CapstoneProject_Correlation Matrix.pdf')
cov_matrix = data_df_num.corr()
plt.figure(figsize = (8, 6))

sns.heatmap(cov_matrix, annot = True, linewidth = 0.8)
plt.title("Covariance Heatmap of Features for Crop Recommendation")
plt.show()
ax = sns.heatmap(cov_matrix, annot = True, linewidth = 0.8)
ax.figure.savefig('CapstoneProject_CovarianceMatrix.pdf')

# Visualize Humidity/Rainfall relationship
plt.figure(figsize = (8, 6))
plt.scatter(data_df["humidity"], data_df["rainfall"], alpha = 0.5)
plt.xlabel("humidity")
plt.ylabel("rainfall")
plt.title("Scatter Plot - humidity vs. rainfall")
plt.grid(True)

# Visualize distributions of Features, with Histograms
data_df_num.hist(figsize = (8, 6), bins = 20, edgecolor = "black")
plt.suptitle("Numeric Features")
ax = plt.suptitle("Distribution of Numeric Features (Histograms)")
ax.figure.savefig('CapstoneProject_HistPlot.pdf')

# Visualize distributions of Features, with VIOLIN Plots
plt.figure(figsize = (8, 6))
sns.violinplot(data = data_df_num, orient = "vertical")
plt.title("Violin Plots of numeric features")
plt.xticks(rotation = 45)
ax = sns.violinplot(data = data_df_num, orient = "vertical")
ax.figure.savefig('CapstoneProject_ViolinPlots.pdf')

# Visualize distributions of Features, with BoxPlots
plt.figure(figsize = (8, 6))
sns.boxplot(data = data_df_num)
plt.title("Box Plots of numeric features")
plt.xticks(rotation = 45)
ax = sns.boxplot(data = data_df_num)
ax.figure.savefig('CapstoneProject_BoxPlots.pdf')

# --- multiple ML models employed: hyperparameter tuning using GridSearch and Cross validation --
# Explored several ML algorithms - choosing the appropriate model for a crop recommendation system depends on several factors, including 
# data size, feature characteristics, interpretability needs, and performance requirements. Evaluating and comparing several models enables 
# us to choose the best model.
# Models selected for evaluation include K-Nearest Neighbors, Random Forest, Gradient-Boost, Support-Vector Machine (SVM) and Logistic 
# Regression.

models = {
    "KNNeighbors": {
        "pipeline": Pipeline([
        ("scaler", StandardScaler()),
        ("model", KNeighborsClassifier())   
    ]),
    "params": {
        "model__n_neighbors": [3, 7],
        "model__weights": ["distance"]
    }
},    
    "RandomForest": {
        "pipeline": RandomForestClassifier(random_state = 42),
        "params": {
            "n_estimators": [50, 100],
            "max_depth": [None, 7],
            "min_samples_leaf": [1, 2]
    }
},   
    "Gradient_Boost": {
        "pipeline": GradientBoostingClassifier(random_state = 42),
        "params": {
            "n_estimators": [50, 100],
            "learning_rate": [0.01, 0.1], 
            "max_depth": [2, 7]
    }   
},
    "SVM": {
        "pipeline": Pipeline([
            ("scaler", StandardScaler()),
            ("model", SVC(probability = True, random_state = 42))
    ]),
        "params": {
            "model__C": [0.1, 1],
            "model__gamma": ["scale", 0.1], 
            "model__kernel": ["rbf"]
    }       
},
    "LogRegression": {
        "pipeline": Pipeline([
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(max_iter = 300, random_state = 42))
    ]),
        "params": {
            "model__C": [0.1, 1],
            # "model__penalty": ["10"], 
            "model__solver": ["lbfgs"]
    }               
}
}

# --- Grid Search ---
# Perform grid search to find the best combination of hyperparameters for the models, and to optimize model performance on the data.
# We evaluate combinations of hyperparameters in a structured approach, to enables a fair comparison between the different 
# models considered, improving the confidence in the recommendation system.

c_val = KFold(n_splits = 5, shuffle = True, random_state = 42)

# ---- best model, given the combination of parameters ----
# 
model_results = []
best_model_gridCV = {}

for model_name, model_params in models.items():
    gridCV = GridSearchCV(
        estimator = model_params["pipeline"],
        param_grid = model_params["params"],
        cv = c_val,
        scoring = "accuracy",
        n_jobs = -1,
        refit = True,
        return_train_score = True
    )
    gridCV.fit(X_features, y_target)
    best_model_gridCV[model_name] = gridCV.best_estimator_


# --- model Scores, comparing performance of models ---
#
model_results = []
for model_name, model_params in models.items():
    model_score = cross_val_score(model_params["pipeline"],
                                  X_features, y_target,
                                  cv = c_val,
                                  scoring = "accuracy"
    )
    model_results.append({
        "model": model_name,
        "cv accu (mean)": model_score.mean(),
        "cv accu (std)": model_score.std(),
        "best hyperparam": gridCV.best_params_,
        "best CV accu":  gridCV.best_score_
    })

# -- DataFrame of Results, Summary form --
# 
model_results_df = (
    pd.DataFrame(model_results).sort_values(by = "best CV accu", ascending = False).reset_index(drop = True)
)
print(model_results_df)

# -- select optimal model --
# 
optimal_model_name = model_results_df.iloc[0]["model"]
optimal_model = best_model_gridCV[optimal_model_name]
print("best model", optimal_model_name)

#  --- Recommendation based on optimal model ---
# 
cropRecommend = optimal_model.predict(X_test)
likelihood = optimal_model.predict_proba(X_test)

print("crop Recommend:", cropRecommend[0])
print("% Predict:", likelihood )

# -- Permutation Feature importance: best model from GridSearchCV --
# In applying Feature Importance in the crop recommendation system, we understand which soil, weather variables have 
# the greatest influence on the most suitable crop to grow, and use such insights to guide crop recommendation.
# 
p_importance = permutation_importance(
    estimator = optimal_model,
    X = X_features,
    y = y_target,
    scoring = "accuracy",
    n_repeats = 10,
    n_jobs = -1
)

# To guide selection of Features that are most important in dataset for better prediction, apply Feature Importance and remove 
# data with neglible importance.
# 
p_importance_df = pd.DataFrame({
    "Feature": X_features.columns,
    "Importance mean": p_importance.importances_mean,
    "Importance stdev": p_importance.importances_std
}).sort_values(
    by = "Importance mean",
    ascending = False
)
print(p_importance_df)

# -- Feature Importance: Visualize data & remove data with neglible importance --
# 
plt.figure()
plt.barh(
    p_importance_df["Feature"],
    p_importance_df["Importance stdev"],
    color = "steelblue"
)
plt.title("Permutation Feature Importance")
plt.xlabel("std dev (accuracy) after Permutation")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
plt.savefig("Permutation Importance.png", dpi = 300, bbox_inches = "tight")

# -- Retrain model with Top "n" Features (most important Features) --
# 
features_topPicks = p_importance_df[
                p_importance_df["Importance mean"] > 0.01]["Feature"].tolist()
print(features_topPicks)

X_features_subset = X_features[features_topPicks]
model_topPicks = optimal_model

# -- fit with tuned model ---
# 
model_topPicks.fit(X_features_subset, y_target)

#  -- reRun CrossValidation --
# 
scores_topPicks = cross_val_score(
    model_topPicks,
    X_features_subset,
    y_target,
    cv = c_val,
    scoring = "accuracy"
)
print("topPicks Features CV accuracy", scores_topPicks.mean())

# -- performance comparison by observing metrics of Top Picks -- 
# 
scores_all = cross_val_score(
    optimal_model,
    X_features,
    y_target,
    cv = c_val,
    scoring = "accuracy"
)

# -- Observations: how do the Accuracies compare? ... accuracy up/ down/ same ?
# 
print ("difference in Accuracy", scores_topPicks.mean() - scores_all.mean())

# -- Crop Recommendation System & logic --
# A crop-recommendation system by means of a user-interaction function() that takes in "Input" data from User and outputs a 
# recommended crop with corresponding Confidence score.
# 
def crop_recommender(model, inputData_with_Features):
    pred = model.predict(inputData_with_Features)[0]
    prob = model.predict(inputData_with_Features).max()
    return pred, prob

# -- Confusion Matrix: evaluate accuracy at the "class" level, to help identify problem areas with insights in misclassification, 
# enabling better-informed recommendations.
# The Confusion Matrix in this task provides a detailed view of how well the model predicts each crop class. It shows the number of 
# correct and incorrect predictions for each class.
# Diagonal entries in the Matrix represent crops that were correctly recommended, while off-diagonal entries show those misclassified.
# The Confusion Matrix allows computation of important metrics, including Accuracy, F1-score,  Precision (how often recommended crop(s) 
# are actually suitable), Recall (how many of truly suitable crop(s) correctly recommended). 
# Through these metrics, the Matrix provides transparent evidence of model performance for each crop, and where recommendations are 
# reliable, or where to apply with caution.
# 

y_pred = cross_val_predict(
    optimal_model,
    X_features,
    y_target,
    cv = c_val,
)

# generate Confusion Matrix to see which crops are correctly/ incorrectly classified.
conf_matrix = confusion_matrix(y_target, y_pred)

conf_matrix_df = pd.DataFrame(
    conf_matrix,
    index = y_target.unique(),
    columns = y_target.unique()
)
plt.figure()
sns.heatmap(conf_matrix_df, annot=True, fmt="d", cmap="viridis")
plt.title("Confusion Matrix")
plt.ylabel("True")
plt.xlabel("Predicted")
plt.savefig("Confusion_Matrix_4_Crop_Recommendation.png", dpi=300, bbox_inches = "tight")


# -- metrics for Confusion Matrix --
# 
accuracy = accuracy_score(y_target, y_pred)
class_report = classification_report(y_target, y_pred, target_names = y_target.unique())
print("Classification Report: \n", class_report)

precision = precision_score(y_target, y_pred, average = None)
recall = recall_score(y_target, y_pred, average = None)
F1 = f1_score(y_target, y_pred, average = None)

cmat_metrics_df = pd.DataFrame({
    "Crop": y_target.unique(),
    "Precision": precision,
    "Recall": recall,
    "F1-score": F1
})
    
print(cmat_metrics_df)

# -- Classification Report --
# From the classification report for above data, we observe that the crop recommendation system had a perfect score for recommending
# Rice, Maze, Kidneybeans, Muskmelon, Apple, Coffee (Accuracy = 1, Precision = 1, F1-score =1).
# The system also had a very high score (score >= 0.99) for many other crops, indicating very good recommendation system.
# Furthermore, macro-average = 0.98 (overall unbiased), weighted-average = 0.98 (accounting for actual distributions of the data), 
# accuracy = 0.98 (system correclty recommends 98% of the time) indicate a consistent strong performance across all crop types, 
# The close alignment of the averages indicate unbiased performance of recommendation system with no significant issues that may 
# arise due to unbalanced data. The system generalizes well, and not favoring certain crops types over others.
# With these observations, it is reasonable to conclude the recommendation system can be considered fair for most of the crop types.
# This indicates a robust and fair recommendation system that can deliver accurate, consistent crop recommendation across a borad range of
# weather conditions and soil types.

```
